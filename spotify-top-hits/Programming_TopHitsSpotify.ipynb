# Analysis and Visualization

# First Question: Who are the top 15 artists with the highest cumulative popularity on Spotify from 2000 to 2019?

```python
Artist_Popularity = df.groupby('Artist')['Popularity'].sum().reset_index().sort_values(by='Popularity',ascending = False).head(15)
Artist_Popularity
# Total Popularity for each artist by grouping the data, sorts them in descending order, and selects the top 15 artists, saving the result in Artist_Popularity. It makes sense to check the cumulative score rather than the mean,as we're not searching for one-hit wonders but artists with great presence during this time period.
```

```python
xticks = Artist_Popularity['Artist']
plt.figure(figsize=(10,10))
plt.bar(Artist_Popularity['Artist'],Artist_Popularity['Popularity'],color='#1DB954')
plt.xticks(xticks, xticks,rotation=90)
plt.show()
#Generates a vertical bar chart showing the Popularity of the top 15 artists, with artist names on the x-axis and their popularity values on the y-axis.
```

## 2.2. How has the popularity of Rihanna's song evolved over the years?


```python
rihanna_songs = df[df['Artist'] == 'Rihanna']
popularity_by_year = rihanna_songs.groupby('Year')['Popularity'].mean().reset_index()
#Analyzes Rihanna's songs to determine the average popularity for each release year. Moreover, it filters the data to include only Rihanna's songs, groups them by their release year, and calculates the mean popularity for each group. The result is a new DataFrame, popularity_by_year, which displays the average popularity of Rihanna's songs year by year.
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Popularity', data=popularity_by_year, marker='o', color='#1DB954')
plt.title("Evolution of Popularity in Rihanna's Songs Over the Years", fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Popularity', fontsize=12)
plt.grid(True)
plt.show()
#Generates a line chart showing how the average popularity of Rihanna's songs has evolved over the years.
```

## 2.4. How have the distributions of key audio features (Popularity, Danceability, Energy, Speechiness, Tempo, Instrumentalness, Liveness, Valence, and Acousticness) in Rihanna's songs evolved on Spotify from 2000 to 2019?"


```python
rihanna_songs = df[df['Artist'] == 'Rihanna']
```

```python
features = ['Popularity', 'Danceability', 'Energy', 'Speechiness', 'Tempo',
            'Instrumentalness', 'Liveness', 'Valence', 'Acousticness']
```

```python
plt.figure(figsize=(15, 12))

for i, feature in enumerate(features, 1):
    plt.subplot(3,3, i)  # 3x3 grid dzeni
    feature_by_year = rihanna_songs.groupby('Year')[feature].mean().reset_index()
    sns.lineplot(x='Year', y=feature, data=feature_by_year, marker='o', color='#1DB954')

    plt.title(f'{feature} Over the Years', fontsize=12)
    plt.xlabel('Year', fontsize=10)
    plt.ylabel(feature, fontsize=10)
    plt.grid(True)

plt.tight_layout()
plt.show()
#Creates a line plots to visualize how various features of Rihanna's songs have evolved over the years.
```
